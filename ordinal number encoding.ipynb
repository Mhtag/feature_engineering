```python
import datetime
```

```python
today_date=datetime.datetime.today()
```

```python
today_date
```

```python
today_date-datetime.timedelta(3)
```

```python
# 15 days date

days = [today_date-datetime.timedelta(x) for x in range(0,15)]
```

```python
days
```

```python
import pandas as pd
df = pd.DataFrame(days)
```

```python
df.columns = ['Day']
```

```python
df['Day'].dt.weekday_name
```

```python
df['weekday'] = df['Day'].dt.day_name()
```

```python
df
```

```python
dictionary={'Monday':1,'Tuesday':2,'Wednesday':3,'Thursday':4,'Friday':5,'Saturday':6,'Sunday':7}
```

```python
df['weekday_ordinal'] = df['weekday'].map(dictionary)
```

### Count Or Frequency Encoding

```python
train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data' , header = None,index_col=None)
train_set.head()
```

```python
columns = [1,3,5,6,7,8,9,13,14]
```

```python
train_set=train_set[columns]
```

```python
train_set.head()
```

```python
train_set.columns = ['Employment','Degree','Status','Designation','family_job','Race','Sex','Country','Salary']
```

```python
train_set.head()
```

```python
for col in train_set.columns:
    print(col, 'has: ',len(train_set[col].unique()),'labels')
```

```python
train_set['Country'].value_counts().to_dict()
```

```python
country_map = train_set['Country'].value_counts().to_dict()
```

```python
train_set['country_map'] = train_set['Country'].map(country_map)
```
